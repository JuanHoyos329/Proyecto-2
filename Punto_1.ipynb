```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import binom, expon

# Configuracin inicial
np.random.seed(42)  # para reproducibilidad

# 1. Distribucin Binomial

# Parmetros
n = 20
p = 0.3
size = 150

# Muestra aleatoria
muestra_binomial = np.random.binomial(n=n, p=p, size=size)

# Valores tericos
x_vals = np.arange(0, n+1)
pmf_vals = binom.pmf(x_vals, n, p)

# Estadsticas tericas
media_teorica_b = n * p
desv_teorica_b = np.sqrt(n * p * (1 - p))
cuartiles_teoricos_b = binom.ppf([0.25, 0.5, 0.75], n, p)

# Estadsticas muestrales
media_muestral_b = np.mean(muestra_binomial)
desv_muestral_b = np.std(muestra_binomial)
cuartiles_muestrales_b = np.percentile(muestra_binomial, [25, 50, 75])

# Graficar histograma y PMF
plt.figure(figsize=(10, 4))
sns.histplot(muestra_binomial, stat="probability", bins=n+1, discrete=True, color='skyblue', label="Muestra")
plt.plot(x_vals, pmf_vals, 'o-', color='red', label='PMF terica')
plt.title("Distribucin Binomial (n=20, p=0.3)")
plt.xlabel("x")
plt.ylabel("Probabilidad")
plt.legend()
plt.grid(True)
plt.show()

print("Distribucin Binomial:")
print(f"Media terica: {media_teorica_b:.2f}, media muestral: {media_muestral_b:.2f}")
print(f"Desviacin estndar terica: {desv_teorica_b:.2f}, muestral: {desv_muestral_b:.2f}")
print(f"Cuartiles tericos: {cuartiles_teoricos_b}")
print(f"Cuartiles muestrales: {cuartiles_muestrales_b}")

# 2. Distribucin Exponencial


# Parmetros
lambd = 1.5
scale = 1 / lambd

# Muestra aleatoria
muestra_exponencial = np.random.exponential(scale=scale, size=size)

# Valores tericos
x_vals_exp = np.linspace(0, max(muestra_exponencial), 100)
pdf_vals = expon.pdf(x_vals_exp, scale=scale)

# Estadsticas tericas
media_teorica_e = scale
desv_teorica_e = scale
cuartiles_teoricos_e = expon.ppf([0.25, 0.5, 0.75], scale=scale)

# Estadsticas muestrales
media_muestral_e = np.mean(muestra_exponencial)
desv_muestral_e = np.std(muestra_exponencial)
cuartiles_muestrales_e = np.percentile(muestra_exponencial, [25, 50, 75])

# Graficar histograma y PDF
plt.figure(figsize=(10, 4))
sns.histplot(muestra_exponencial, bins=20, stat="density", color='lightgreen', label='Muestra')
plt.plot(x_vals_exp, pdf_vals, color='darkgreen', label='PDF terica')
plt.title("Distribucin Exponencial ( = 1.5)")
plt.xlabel("x")
plt.ylabel("Densidad")
plt.legend()
plt.grid(True)
plt.show()

print("\nDistribucin Exponencial:")
print(f"Media terica: {media_teorica_e:.2f}, media muestral: {media_muestral_e:.2f}")
print(f"Desviacin estndar terica: {desv_teorica_e:.2f}, muestral: {desv_muestral_e:.2f}")
print(f"Cuartiles tericos: {cuartiles_teoricos_e}")
print(f"Cuartiles muestrales: {cuartiles_muestrales_e}")
```
